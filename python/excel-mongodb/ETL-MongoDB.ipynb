```python
import pandas as pd
import numpy as np
```

# Extraction and Transformation

## Extracting and Transforming Air pollution Dataset version 2022

```python
df_2022=pd.read_excel("data/who_aap_2021_v9_11august2022.xlsx",sheet_name=1)
df_2022.drop(columns='Status',inplace=True)
df_2022.rename(columns={'WHO Country Name':'Country','Version of the database':'Database version (year)',
                       'PM25 temporal coverage (%)':'PM2.5 temporal coverage (%)'}, inplace=True)
```

```python
df_2022.head()
```

## Extracting and Transforming Air pollution Dataset version 2018

```python
df_2018=pd.read_excel("data/aap_air_quality_database_2018_v14-(1).xlsx",sheet_name=1,header=2)
df_2018.rename(columns={'Region':'WHO Region',
                        'iso3':'ISO3',
                        'City/Town':'City or Locality',
                        'Year':'Measurement Year',
                        'Annual mean, ug/m3':'PM10 (μg/m3)',
                        'Temporal coverage':'PM10 temporal coverage (%)',
                        'note on converted PM10':'Note on Converted PM10',
                        'Annual mean, ug/m3.1':'PM2.5 (μg/m3)',
                        'Temporal coverage.1':'PM2.5 temporal coverage (%)',
                        'note on converted PM2.5':'Note on Converted PM2.5',
                        'Reference for air quality':'Reference',
                       },inplace=True)
df_2018.drop(columns='status',inplace=True)
```

```python
df_2018['WHO Region'].replace({r'(\(.*)':'',}, regex=True, inplace=True)
```

```python
df_2018.tail()
```

## Extracting and Transforming Air pollution Dataset version 2016

```python
df_2016=pd.read_excel("data/who-aap-database-may2016.xlsx",sheet_name=1,header=2)
df_2016.head()
```

```python
df_2016[['PM10 temporal coverage (%)','PM2.5 temporal coverage (%)']]=df_2016['Temporal coverage'].str.split(pat=";",expand=True)
```

```python
df_2016.rename(columns={'Region':'WHO Region',
                        'iso3':'ISO3',
                        'City/Town':'City or Locality',
                        'Year':'Measurement Year',
                        'Annual mean, ug/m3':'PM10 (μg/m3)',
                        'note on converted PM10':'Note on Converted PM10',
                        'Annual mean, ug/m3.1':'PM2.5 (μg/m3)',
                        'note on converted PM2.5':'Note on Converted PM2.5',
                        'for air quality':'Reference',
                        'Number and type of monitoring stations (PM10)':'Number and type of monitoring stations'
                       },inplace=True)
```

```python
df_2016['Temporal coverage'].unique()
```

```python
df_2016.columns
```

```python
df_2016.drop(columns={'Year.1','ID WHO city','Temporal coverage','Number and type of monitoring stations (PM2.5)'},inplace=True)
```

```python
df_2016['PM2.5 temporal coverage (%)']=df_2016['PM2.5 temporal coverage (%)'].replace({r'.*(?=[<>])':'',r'(?<=%).*':'',' ':'','NA':np.NaN}, regex=True)
df_2016['PM10 temporal coverage (%)']=df_2016['PM10 temporal coverage (%)'].replace({r'.*(?=[<>])':'',r'(?<=%).*':'',' ':'','NA':np.NaN}, regex=True)
```

```python
df_2016.head()
```

## Extracting and Transforming Air pollution Dataset version 2014

```python
df_2014=pd.read_excel("data/aap_pm_database_may2014.xls",sheet_name=1,header=[1])
```

```python
df_2014.rename(columns={'Region':'WHO Region',
                        'PM10':'PM10 (μg/m3)',
                        'PM10.1':'Measurement Year',
                        'City/station':'City or Locality',
                        'PM 2.5':'PM2.5 (μg/m3)',
                        'note on converted PM10':'Note on Converted PM10',
                        'Annual mean, ug/m3.1':'PM2.5 (μg/m3)',
                        'note on converted PM2.5':'Note on Converted PM2.5',
                        'for air quality':'Reference',
                        'of monitoring stations':'Number and type of monitoring stations'
                       },inplace=True)
```

```python
df_2014.drop(columns={'Unnamed: 13','Unnamed: 14','PM2.5','Subregion'},inplace=True)
```

```python
df_2014.tail(11)
```

```python
df_2014.drop(df_2014.tail(10).index,inplace=True)
```

## Extracting and Transforming Air pollution Dataset version 2011

```python
df_2011_1=pd.read_excel("data/oap_database-(1).xls",sheet_name=1,header=[1])
df_2011_2=pd.read_excel("data/oap_database-(1).xls",sheet_name=3,header=[1])
```

```python
df_2011=df_2011_2.merge(df_2011_1, on=('Country','City'), how='left')
```

```python
df_2011.head()
```

```python
df_2011.drop(index=[0],inplace=True)
```

```python
df_2011.columns
```

```python
df_2011.rename(columns={'Region_x':'WHO Region',
                        'City':'City or Locality',
                        'Annual mean PM10':'PM10 (μg/m3)',
                        'Annual mean PM2.5':'PM2.5 (μg/m3)',
                        'Year_x':'Measurement Year',
                        'Number and location_x':'Number and type of monitoring stations',
                        'Temporal coverage_x':'PM10 temporal coverage (%)',
                        'Temporal coverage_y':'PM2.5 temporal coverage (%)',
                        'Reference_x':'Reference',},inplace=True)           
```

```python
df_2011.drop(columns={'Region_y', 'Year_y', 'Number and location_y','Reference_y'},inplace=True)
```

```python
df_2011.columns
```

## Combining dataframes

```python
df=pd.concat([df_2022,df_2018,df_2016,df_2014,df_2011],ignore_index=True)
```

```python
df.info()
```

```python
df[['Country','City or Locality','Measurement Year']].drop_duplicates(ignore_index=True,keep='first',inplace=True)
```

## Cleaning columns - WHO Region , ISO3 , Country

```python
grouping=pd.read_csv('data/WHO Country Grouping.csv')
```

```python
region_iso_dict=pd.DataFrame(grouping[['DisplayString','WHO_REGION','ISO']].set_index('DisplayString').to_dict()).to_dict('index')
```

```python
region_iso_dict['United Kingdom']={'WHO_REGION': 'Europe', 'ISO': 'GBR'}
region_iso_dict['Turkey']={'WHO_REGION': 'Europe', 'ISO': 'TUR'}
```

```python
df['Country'].replace({'Bolivia, Plurinational States of':'Bolivia (Plurinational State of)',
                      'The former Yugoslav Republic of Macedonia':'Yugoslavia, Former',
                       'Venezuela, Bolivarian Republic of':'Venezuela (Bolivarian Republic of)',
                       'Venezuela':'Venezuela (Bolivarian Republic of)',
                       'Czech Republic':'Czechia',
                       'Unit. Rep. of Tanzania':'United Republic of Tanzania',
                       'Iran':'Iran (Islamic Republic of)',
                       'Israel ':'Israel'                       
                      },inplace=True)
```

```python
df['WHO Region']=df.apply(lambda x: region_iso_dict[x.Country]['WHO_REGION'], axis=1)
df['ISO3']=df.apply(lambda x: region_iso_dict[x.Country]['ISO'], axis=1)
```

## Sorting based on column - City or Locality

```python
df=df.sort_values(by=['Country','City or Locality'])
```

```python
df.reset_index(drop=True)
```
